```python
fig=plt.figure(figsize=(16,9))
plt.title("NUMBER OF FRAUDS IN EACH STATE IN 2019-20",size=25)
ax=sns.barplot(data=df,y="Number  _of_frauds_2019_20\n",x="NAME_OF_THE_STATES",linewidth=1,edgecolor="red")
plt.xticks(rotation=100)
ax.bar_label(ax.containers[0],size=12)
```

```python
#pivot table
statewise=pd.pivot_table(df,values=["Extent_of_loss_2019_20","Amount_Involved_2019_20","Number  _of_frauds_2019_20\n"],index="NAME_OF_THE_STATES",aggfunc=max)
statewise.style.background_gradient(cmap="cubehelix")
```

```python
#pivot table
statewise=pd.pivot_table(df,values=["Extent_of_loss_2019_20","Amount_Involved_2019_20","Number  _of_frauds_2019_20\n"],index="NAME_OF_THE_STATES",aggfunc=max)
statewise.style.bar(cmap="cubehelix")
```

```python
top10_states=df.groupby(by='NAME_OF_THE_STATES').max()[['Extent_of_loss_2019_20','Amount_Involved_2019_20']].sort_values(by=['Extent_of_loss_2019_20'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 States in Fraud amount involved in 2019-20",size=25)
ax=sns.barplot(data=top10_states.iloc[:10],x="NAME_OF_THE_STATES",y="Amount_Involved_2019_20",linewidth=2,edgecolor="red")
plt.xticks(rotation=100)
ax.bar_label(ax.containers[0],size=12)
```
